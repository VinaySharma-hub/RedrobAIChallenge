```python
from collections import Counter
class Vocabulary:
    def __init__(self,min_freq=2):
        self.min_freq=min_freq
        self.special_tokens=['<PAD>', '<BOS>', '<EOS>', '<UNK>']
        self.stoi={tok:idx for idx,tok in enumerate(self.special_tokens)}
        self.itos={idx:tok for idx,tok in enumerate(self.special_tokens)}
    def build_vocab(self,caption):
        counter=Counter()
        for captions in caption:
            counter.update(str(captions).lower().split())
        idx=len(self.special_tokens)
        for word,freq in counter.items():
            if freq>=self.min_freq:
                self.stoi[word]=idx
                self.itos[idx]=word
                idx+=1 
```
